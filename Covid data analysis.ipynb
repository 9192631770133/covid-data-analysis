# Recovered countries name in descending order

```python
df.groupby("country").Recovered.mean().sort_values(ascending=False) # higher recovered country is brazil
```

# Top 10 Covid Recovered countries and its representation

```python
df.groupby("country").Recovered.mean().sort_values(ascending=False).nlargest(10)
```

```python
df.groupby("country").Recovered.mean().sort_values(ascending=False).nlargest(10).plot(kind="bar")
```

# Top 10 countries of Covid Deaths cases and its representation

```python
df.groupby("country").Deaths.mean().sort_values(ascending=False).nlargest(10)
```

```python
df.groupby("country").Deaths.mean().sort_values(ascending=False).nlargest(10).plot.pie(autopct="%.2f%%")
plt.title("Deaths ratio")
```

```python
df.groupby("country").Confirmed.mean().sort_values(ascending=False).nlargest(10).plot.pie(autopct="%0.2f%%")
plt.title("confirmed covid case ratio")
```

# Representing the confirmed cases with their avarage and middle value

```python
confirmed_mean=df["Confirmed"].mean()
confirmed_median=df["Confirmed"].median()
sns.distplot(df["Confirmed"])
plt.axvline(confirmed_median,color="red")
plt.axvline(confirmed_median,color="yellow")
plt.show()
```
